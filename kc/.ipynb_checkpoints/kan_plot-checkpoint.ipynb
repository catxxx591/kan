```python
df = kc3.getsqldata()
df
```

```python
BB = df.loc[df['type']== '装甲空母']
CC = df.loc[df['type']== '正規空母']
BB = pd.concat([BB,CC])
```

```python
BB.columns
```

```python
BB = BB.loc[BB['top']=='True']
BBclass = BB.groupby('class').mean().round()
BBclass
```

```python
BBclass.dtypes
```

```python
BBclass = BBclass.T
BBclass
```

```python
n= 0
BBclass['mean']=1
while n < len(BBclass['mean']):
    BBclass['mean'][n]=BBclass.loc[BBclass.index[n]].mean().round()
    n += 1
BBclass
```

```python
BBclass.drop(['fire','torpedo','aa','asw','hp','armor','evasion','los','luck'],inplace=True)
```

```python
BBclass2 = BBclass.T.apply(lambda x:x.apply(lambda y:y-x[-1]))
BBclass2
```

```python
BBclass2.T[['mean','翔鶴型','Lexington級','Essex級','加賀型']].plot(kind='line',alpha=0.75)
```

```python
name = '五十鈴改二'
shiptype = list(df.loc[df['name']==name,'type'])[0]
shipclass = list(df.loc[df['name']==name,'class'])[0]
shiptype
```

```python
BBclass
```

```python
BB = df.loc[df['type']== shiptype]
BB = BB.loc[BB['top']=='True']
BBclass = BB.groupby('class').mean().round()
#BBtop = BB.loc[BB['top']=='True']#
#BBtopclass = BBtop.groupby('class').mean().round()#
BBclass = BBclass.T
#BBtopclass = BBtopclass.T#
n= 0
BBclass['mean']=1
while n < len(BBclass['mean']):
    BBclass['mean'][n]=BBclass.loc[BBclass.index[n]].mean().round()
    n += 1
BBclass
BBclass.drop(['fire','torpedo','aa','asw','hp','armor','evasion','los','luck'],inplace=True)
BBclass2 = BBclass.T.apply(lambda x:x.apply(lambda y:y-x[-1]))
#BBclass2.T[['mean',shipclass]].plot(kind='line',alpha=0.75)
#BBclass2.columns.apply(lambda x : x.split("_")[0])
coldict = {'fire_max':'火力','torpedo_max':'雷裝','aa_max':'對空','asw_max':'對潛','hp_max':'耐久','armor_max':'裝甲','evasion_max':'迴避','carry':'搭載','los_max':'索敵','luck_max':'運'}
#BBclass2.rename(columns=lambda x: x.split("_")[0], inplace=True)
BBclass2.rename(columns=coldict, inplace=True)
BBclass2 = BBclass2.T
```

```python
from matplotlib.font_manager import FontProperties
import matplotlib.pyplot as plt 
myfont = FontProperties(fname='C:\Windows\Fonts\simhei.ttf', size=12)
plt.plot(BBclass2.index,BBclass2['mean']) 

plt.bar(BBclass2.index,BBclass2[shipclass])

#plt.title("聲量圖",fontproperties=myfont) 
#plt.ylabel("文章數量",fontproperties=myfont) 
#plt.xlabel("時間",fontproperties=myfont)  
plt.show()
```
