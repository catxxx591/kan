```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import re
import requests
import time
import random
```

```python
def search(i='',count=500):
    p=1
    itemdict={}
    while True:
        url = 'https://find.ruten.com.tw/s/?p='+str(p)+'&q='+i
        print('第'+str(p)+'頁',end="")
        headers={'User-Agent':'Mozilla/5.0 (iPhone; CPU iPhone OS 7_1_2 like Mac OS X) App leWebKit/537.51.2 (KHTML, like Gecko) Version/7.0 Mobile/11D257 Safari/9537.53'}

        res = requests.get(url,headers=headers)
        res.encoding = 'utf-8'
        soup = BeautifulSoup(res.text,'html.parser')
        itemprice = soup.find_all('body')[0].find_all('div',{'class':'item-price text-price rt-text-price js-comma-format'})
        itemname = soup.find_all('body')[0].find_all('div',{'class':'item-name-text'})
        if len(itemname)==0:
            break
        
        n=0        
        while n < len(itemprice):
#            print(itemname[n].text)
            itemdict[itemname[n].text] = itemprice[n].text
            if len(list(itemdict.keys())) >= count:
                break
            n += 1
        if len(list(itemdict.keys())) >= count:
            break
        time.sleep(5.5+(random.randint(-2,2)/5))

        p += 1
    return itemdict

#soup = BeautifulSoup(res.text,'html.parser')
#soup = BeautifulSoup(res.decode('utf-8'),'html.parser')
#soup = BeautifulSoup(res.text.decode(PYTHONIOENCODING=utf-8),'html.parser')
#soup = BeautifulSoup(res.read().decode('utf-8'),'html.parser')
#soup = BeautifulSoup(res.text.decode('utf-8'),'html.parser')
#soup

#print("encoding: %s" % res.encoding)
#print("content: \n%s" % res.text)
#res.text
```
